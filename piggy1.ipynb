```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import netCDF4 as nc
```

```python
ds = nc.Dataset('20400101.aijh2E213SSP119aF40oQ40.nc4')
print(ds)
```

```python
lat = ds.variables['lat'][:]
lon = ds.variables['lon'][:]
time = ds.variables['time'][:]
```

```python
fig, [ax,ax2] = plt.subplots(figsize=(10,5), nrows=1, ncols=2)
cm = ['RdBu_r', 'viridis']

im = ax.contourf(ds['T2M'][0,:,:])
fig.colorbar(im,ax=ax)

im = ax2.contourf(ds['PRECTOT'][0,:,:])
fig.colorbar(im,ax=ax2)
```

```python
TM2_np = np.array(ds['T2M'])
PRECTOT_np = np.array(ds['PRECTOT'])
print(TM2_np.shape,PRECTOT_np.shape)
```

```python
#average over time axis
TM2_np_av = TM2_np.mean(axis=0)
PRECTOT_np_av = PRECTOT_np.mean(axis=0)
```

```python
fig, [ax,ax2] = plt.subplots(figsize=(10,5), nrows=1, ncols=2)
cm = ['RdBu_r', 'viridis']

im = ax.contourf(TM2_np_av)
fig.colorbar(im,ax=ax)

im = ax2.contourf(PRECTOT_np_av)
fig.colorbar(im,ax=ax2)
```
